=========================Part I - The Node class===========================

```python

class Node:
    def __init__(self, name):
        str_name = str(name)
        self.name = str_name
        neighbors ={}
        self.neighbors = neighbors
        self.num_neighbors = len(neighbors)
    def __str__(self):
        return "Node name is:" + str(self.name) + " and his neighbors are " + str(self.neighbors)
    def __len__(self):
        return self.num_neighbors

    def __eq__(self, other):
        if self.name == other.name :
            ret_val = True
        else :
            ret_val = False
        return ret_val
    def __ne__(self, other):
        if self.name != other.name :
            ret_val = True
        else :
            ret_val = False
        return ret_val
    def add_neighbor(self, name, weight=1):
        str_name = str(name)
        if name not in self.neighbors.keys() and str_name != self.name:
            self.neighbors[str_name] = weight
            self.num_neighbors+=1
    def remove_neighbor(self, name):
        str_name = str(name)
        del self.neighbors[str_name]
        print 'Node',name,'was deleted'

    def is_pointing_at(self, name):
        str_name = str(name)
        if str_name in self.neighbors.keys() :
            ret_val = True
        else :
            ret_val = False
        return ret_val

    def get_weight(self, name):
        if self.get_neighbors(name):
            ret_val = self.neighbors[name]
        else :
            return
        return ret_val

    def get_neighbors(self,name):
        return name in self.neighbors
    def is_isolated(self,name):
        for key in self.nodes.values():
            if key.is_pointing_at(name):
                return  False
        return True
```

=========================Part II - The Graph class=========================== 

```python
from collections import namedtuple
PathInfo = namedtuple('PathInfo', ('nodes', 'edges', 'costs', 'total_cost'))

class OlaError(Exception):
    '''Base class for Ola's errors.'''

class NoPathError(OlaError):
    '''Raised when a path can't be found to a specified node.'''

class Graph:
    def __init__(self, name, nodes=[]):
        self.name = name
        self.nodes={}
        for node in nodes:
            self.add_node(node)

    def __str__(self):
        return "Graph name is: " + self.name + " and it has " + str(len(self.nodes))+"\n"+ "\n".join([str(self.nodes[i]) for i in self.nodes.keys()])

    def __len__(self):
        return len(self.nodes)
    def __contains__(self, key):
        if isinstance(key, basestring) and self.nodes.has_key(key):
            return True
        if isinstance(key, Node) and self.nodes.has_key(key.name):
            return True
        print ('Error')
        return False
    def __getitem__(self, name):
        return self.nodes[name]

    def __add__(self, other):
        new_graph=Graph(self.name,self.nodes.values())
        validate_add = False
        for k,v in other.nodes.items():
            if new_graph.nodes.has_key(k):
                validate_add = True
            new_graph.add_node(v)
        if validate_add:
            return new_graph
        else:
            raise KeyError('Failed common node validation')

    def add_node(self, node):
        if isinstance(node,Node):
            if self.nodes.has_key(node.name):
                new_node_neighbors={}
                for k, v in node.neighbors.items():
                    new_node_neighbors[k] = v
                for k, v in self.nodes[node.name].neighbors.items():
                    new_node_neighbors[k] = v
                self.nodes[node.name].neighbors = new_node_neighbors
            else:
                self.nodes[node.name]=node
        else:
            raise KeyError ('Invalid data type')

    def remove_node(self, name):
        if self.nodes.has_key(name):
            del self.nodes[name]

    def is_edge(self, frm_name, to_name):
        if self.nodes.has_key(frm_name)==False:
            return False
        return self.nodes[frm_name].is_pointing_at(to_name)

    def add_edge(self, frm_name, to_name, weight=1):
        if self.nodes.has_key(frm_name) and self.nodes.has_key(to_name):
            if not self.nodes[frm_name].is_pointing_at(to_name):
                if not self.nodes[frm_name] == self.nodes[to_name]:
                    self.nodes[frm_name].add_neighbor(to_name, weight)

    def remove_edge(self, frm_name, to_name):
        if self.nodes.has_key(frm_name):
            self.nodes[frm_name].remove_neighbor(to_name)


    def get_edge_weight(self, frm_name, to_name):
        if self.nodes.has_key(frm_name) and self.nodes.has_key(to_name):
            if self.nodes[frm_name].is_pointing_at(to_name):
                return self.nodes[frm_name].get_weight(to_name)
        return None

    def get_path_weight(self, path):
        #validation
        import collections
        if isinstance(path, collections.Iterable) or len(path)< 2:
            return None
        for path_elem in path:
            if not self.nodes.has_key(path_elem):
                return None

        prev_node = path[0]
        path_weight = 0
        for i in range(0,len(path)):
            if self.is_edge(prev_node, str(path[i])):
                path_weight += self.get_edge_weight(prev_node, str(path[i]))
            else:
                return
            prev_node = path[i]
            return path_weight

    def is_reachable(self, frm_name, to_name, n=0):
        if n > len(self) or frm_name not in self or to_name not in self:
            return False
        elif str(frm_name) == str(to_name) or self.is_edge(frm_name, to_name) or True in [self.is_reachable(i, to_name, n + 1) for i in self[frm_name].get_neighbors()]:
            return True
        else:
            return False
    def is_isolated(self,name):
        for key in self.nodes.values():
            if key.is_pointing_at(name):
                return  False
        return True

    def shortest_paths(self, s, d):
        from itertools import count
        from heapq import heappush, heappop
        counter = count()
        costs = {s: 0}
        predecessors = {s: (None, None, None)}
        visit_queue = [(0, next(counter), s)]
        visited = set()

        while visit_queue:
            cost_of_s_to_u, _, u = heappop(visit_queue)
            if u == d:
                break

            if u in visited:
                continue

            visited.add(u)
            neighbors = self[u].neighbors

            if not neighbors:
                # u has no outgoing edges
                continue

            for v in neighbors.keys():
                if v in visited: # Don't backtrack to nodes that have already been visited.
                    continue

                cost_of_e= neighbors[v]
                cost_of_s_to_u_plus_cost_of_e = cost_of_s_to_u + cost_of_e

                if v not in costs or costs[v] > cost_of_s_to_u_plus_cost_of_e:
                    costs[v] = cost_of_s_to_u_plus_cost_of_e
                    predecessors[v] = (u, e, cost_of_e)
                    heappush(visit_queue, (cost_of_s_to_u_plus_cost_of_e, next(counter), v))

        if d is not None and d not in costs:
            raise NoPathError('Could not find a path from {0} to {1}'.format(s, d))
        return predecessors

    def extract_shortest_path_from_predecessor_list(self, predecessors, d):
        nodes = [d]
        edges = []
        costs = []
        u, e, cost = predecessors[d]
        while u is not None:
            nodes.append(u)
            edges.append(e)
            costs.append(cost)
            u, e, cost = predecessors[u]
        nodes.reverse()
        edges.reverse()
        costs.reverse()
        total_cost = sum(costs)
        return PathInfo(nodes, edges, costs, total_cost)

    def find_shortest_path(self, frm_name, to_name):
        predecessors = self.shortest_paths(frm_name, to_name)
        shortest_path = self.extract_shortest_path_from_predecessor_list(predecessors, to_name)
        min_weight = shortest_path.total_cost
        min_weight_path = [self.nodes[nn] for nn in shortest_path.nodes]
        return (min_weight_path, min_weight)
```

```python
def parse_date(dt, dtf):
    real_dt=None
    try:
        real_dt = datetime.strptime(dt, dtf)
    except ValueError as e:
        print '''Bad format''',e
    return real_dt

from datetime import datetime
import time
known_regions = ['Center','North','East','West','South']

def create_travel_graph(file, graph_name, dtf='%d/%m/%Y %H %M'):
    gr = Graph(graph_name)
    with open(file, 'r') as csvfile:
        cnt = 0
        denominator_dict = {}
        for line in csvfile:
            if cnt > 1:
                split = line.split(',')
                if len(split) == 4:
                    reg1,dt1,reg2,dt2 = split
                    if reg1 in known_regions and reg2 in known_regions:
                        dt1 = parse_date(dt1.replace('h',' ').replace('m',' ').strip(), dtf)
                        dt2 = parse_date(dt2.replace('h',' ').replace('m',' ').strip(), dtf)
                        if dt1 and dt2:
                            d1_ts = time.mktime(dt1.timetuple())
                            d2_ts = time.mktime(dt2.timetuple())
                            minutes_travel = int(d2_ts - d1_ts) / 60 * 60  # They are now in seconds, subtract and then divide by 60 to get minutes.
                            # print 'from',reg1,'to',reg2,'takes',minutes_travel
                            if not reg1 in gr:
                                gr.add_node(Node(reg1))
                            if not reg2 in gr:
                                gr.add_node(Node(reg2))

                            if not gr.is_edge(reg1, reg2):
                                gr.add_edge(reg1, reg2, minutes_travel)
                                denominator_dict[(reg1,reg2)] = 1
                            else:
                                prev_average = gr.get_edge_weight(reg1, reg2)
                                prev_denominator = denominator_dict[(reg1,reg2)]
                                new_denominator = prev_denominator +1
                                new_average = (prev_average*prev_denominator + minutes_travel)/new_denominator
                                gr.add_edge(reg1, reg2, new_average)
                                denominator_dict[(reg1, reg2)] = new_denominator
                                print 'New data is ',minutes_travel,' and average for',reg1,reg2,'is',new_average,'and count is',new_denominator
                        else:
                            print 'Bad time format at line',cnt, 'dt1',dt1,'dt2',dt2
                    else:
                        print 'Bad region at line',cnt
                else:
                    print 'Bad split at line',cnt
            cnt +=1

        print 'read', str(cnt-1), 'lines'
    return gr
```
